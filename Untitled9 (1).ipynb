```python
import base64

def convert_to_base64(file_path):
    with open(file_path, "rb") as f:
        encoded = base64.b64encode(f.read()).decode('utf-8')
    return encoded

def convert_folder_to_base64(folder_path):
    base64_data = {}
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_path = os.path.join(root, file)
            b64 = convert_to_base64(file_path)
            base64_data[file_path] = b64
    return base64_data


img_path = os.path.join(extract_folder, "img")
docs_path = os.path.join(extract_folder, "docs")
pdf_path = os.path.join(extract_folder, "pdf")


img_b64 = convert_folder_to_base64(img_path)
docs_b64 = convert_folder_to_base64(docs_path)
pdf_b64 = convert_folder_to_base64(pdf_path)

print("‚úÖ Images converted:", list(img_b64.keys()))
print("‚úÖ Docs converted:", list(docs_b64.keys()))
print("‚úÖ PDFs converted:", list(pdf_b64.keys()))
```

```python
from google.colab import files
import zipfile
import os
import base64


uploaded = files.upload()


zip_name = list(uploaded.keys())[0]
with zipfile.ZipFile(zip_name, 'r') as zip_ref:
    zip_ref.extractall()


root_folders = [f for f in os.listdir() if os.path.isdir(f) and not f.startswith(".")]
print("üìÅ Extracted folders:", root_folders)
extract_folder = root_folders[0]


img_path = os.path.join(extract_folder, "img")
docs_path = os.path.join(extract_folder, "docs")
pdf_path = os.path.join(extract_folder, "pdf")


def convert_to_base64(folder):
    encoded_files = {}
    if not os.path.exists(folder):
        print(f"‚ùå Folder not found: {folder}")
        return {}
    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)
        with open(file_path, "rb") as f:
            encoded = base64.b64encode(f.read()).decode('utf-8')
            encoded_files[filename] = encoded
    return encoded_files


img_b64 = convert_to_base64(img_path)
docs_b64 = convert_to_base64(docs_path)
pdf_b64 = convert_to_base64(pdf_path)


if img_b64:
    print("\nüì∏ Sample Base64 of an Image:")
    print(list(img_b64.values())[0][:200], "...")

if docs_b64:
    print("\nüìÑ Sample Base64 of a Doc:")
    print(list(docs_b64.values())[0][:200], "...")

if pdf_b64:
    print("\nüìï Sample Base64 of a PDF:")
    print(list(pdf_b64.values())[0][:200], "...")


if img_b64:
    with open("decoded_image.png", "wb") as f:
        f.write(base64.b64decode(list(img_b64.values())[0]))
if docs_b64:
    with open("decoded_doc.docx", "wb") as f:
        f.write(base64.b64decode(list(docs_b64.values())[0]))
if pdf_b64:
    with open("decoded_pdf.pdf", "wb") as f:
        f.write(base64.b64decode(list(pdf_b64.values())[0]))

print("\n‚úÖ Decoded files saved: decoded_image.png, decoded_doc.docx, decoded_pdf.pdf")
```
